# STEP 3 Splitiing training and testing data, scaling the data, performing logistic Regression

```python
X = df_balanced.drop(['Time','Class'],axis=1)
y = df_balanced['Class']
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state =42)
# Display the shape of the training and testing sets
print("Training set shape - X_train:", X_train.shape, "y_train:", y_train.shape)
print("Testing set shape - X_test:", X_test.shape, "y_test:", y_test.shape)
```

```python
scaler= StandardScaler()

train_scaled = scaler.fit_transform(X_train)
test_scaled = scaler.transform(X_test)
```

```python
log_reg_model = LogisticRegression(max_iter=1000)

log_reg_model.fit(X_train, y_train)
```

# STEP 4 Displaying training and testing evaluation metrics, and feature importance

```python
predicted_training = log_reg_model.predict(X_train)

accuracy = accuracy_score(predicted_training,y_train)*100
precision = precision_score(predicted_training,y_train)*100
recall = recall_score(predicted_training,y_train)*100
f1 = f1_score(predicted_training,y_train)*100

# Output the evaluation metrics
print("Output for training data")
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```

```python
predicted_testing = log_reg_model.predict(X_test)

accuracy = accuracy_score(predicted_testing,y_test)*100
precision = precision_score(predicted_testing,y_test)*100
recall = recall_score(predicted_testing,y_test)*100
f1 = f1_score(predicted_testing,y_test)*100

# Output the evaluation metrics
print("Output for training data")
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```

Now lets explore the feature importance

```python
coefficients = log_reg_model.coef_[0]

importances = np.abs(coefficients)

indices = np.argsort(importances)[::-1]

plt.figure(figsize=(12, 6))
plt.title("Feature Importances - Logistic Regression")
plt.bar(range(len(indices)), importances[indices], align="center")
plt.xticks(range(len(indices)), indices)
plt.xlabel("Features")
plt.ylabel("Importance (Absolute Coefficient)")
plt.tight_layout()
plt.show()
```

Here the most important feature is V13

## Insights from the analysis

Our logistic regression model achieved an accuracy of 95.61% in detecting credit card fraud, with V13 identified as a key feature. This suggests that V13 significantly contributes to distinguishing fraudulent transactions, highlighting its potential role in capturing critical fraud patterns.
